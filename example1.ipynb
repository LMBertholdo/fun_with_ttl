## Libraries

```python
import pandas as pd 
import numpy as np
```

## Reading the filtered pcap file into a dataframe

```python
df = pd.read_csv('pcap.txt', error_bad_lines=False, sep=';')
```

## Quick preview of the dataframe

```python
df.head()
```

## Number of packets per source IP address

```python
summary = df['ip.src'].value_counts()
summary
```

## Group by source IP address

```python
summary_max = df.groupby(['ip.src'])['ip.ttl'].max()
summary_max
```

```python
summary_min = df.groupby(['ip.src'])['ip.ttl'].min()
summary_min
```

```python
df_summary = pd.concat([summary, summary_min, summary_max], axis=1)
df_summary.columns=['num_packets','min_ttl','max_ttl']
df_summary
```

```python
df_summary['diff']=df_summary['max_ttl']-df_summary['min_ttl']
df_summary
```

## One line of code for calculating ALL stats while aggregating

```python
# https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
df_summary2 = df.groupby(['ip.src'])['ip.ttl'].agg(['count', 'min', 'max','mean','std', np.ptp])
df_summary2
```

#### Resetting Index

```python
df_summary2 = df_summary2.reset_index()
df_summary2
```

## Libraries

```python
import pandas as pd 
import numpy as np
```

## Reading the filtered pcap file into a dataframe

```python
df = pd.read_csv('pcap.txt', error_bad_lines=False, sep=';')
```

## Quick preview of the dataframe

```python
df.head()
```

```python
df['ip.src_first'] = df['ip.src'].apply(lambda x: x.split(',')[0])
df
```

```python
df['ip.ttl_first'] = df['ip.ttl'].apply(lambda x: int(x.split(',')[0]))
df
```

```python
df_summary = df.groupby(['ip.src_first'])['ip.ttl_first'].agg(['count', 'min', 'max','mean','std','median', np.ptp]).reset_index()
df_summary
```

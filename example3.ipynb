## Libraries

```python
import pandas as pd 
import numpy as np
```

## Reading the filtered pcap file into a dataframe

```python
df = pd.read_csv('pcap.txt', error_bad_lines=False, sep=';')
```

## Quick preview of the dataframe

```python
df.head()
```

```python
df_summary = df.groupby(['ip.src'])['ip.ttl'].agg(['count', 'min', 'max','mean','std','median', np.ptp]).reset_index()
df_summary
```
